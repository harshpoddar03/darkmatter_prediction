```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

x = np.load('train/sphere/10.npy')
x = x.squeeze(0)

# Assuming `image` is your loaded NumPy array from the .npy file
image = x
# Convert image to grayscale if it's not already
if image.ndim == 3:
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
else:
    image_gray = image

# Ensure the image is of type uint8
image_gray = np.uint8(image_gray)

# Apply Canny edge detection
edges = cv2.Canny(image_gray, 10, 20)

# Plotting the original image and the edges
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(x, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Edges')
plt.imshow(edges, cmap='gray')
plt.axis('off')

plt.show()
print('Original image shape:', x.ndim)
print('Image shape:', x.shape)
print('Data type:', image_gray.dtype)
print('Edges shape:', edges.shape)
```
